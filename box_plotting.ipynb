```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


with open('number_of_routing_decsn_2.txt') as f:
    lines = f.readlines()

acc_dict = {}
acc_list = []
substr = 'router'
values_list = []
router_cnt = -1
for line in lines:
  if substr in line:
    continue
  router_cnt += 1
  values = line.split(",")
  values = values[0: len(values)-1]
  values = np.array(values)
  values = values.astype(int)
  print(len(values))
  for val in values:
    tuple_ = [router_cnt, val]
    values_list.append(tuple_)

  values_list_np = np.array(values_list)

number_of_routing_decisions = values_list_np.copy
plt.rcParams["figure.figsize"] = (16,12)
dataframe=pd.DataFrame(data=values_list_np,columns = ['src','number_of_routing_decisions']) 
ax = sns.boxplot(x="src", y="number_of_routing_decisions", data=dataframe)
dataframe.boxplot(by ='src', column =['number_of_routing_decisions'], grid = False)
plt.show(block=True);

print(dataframe)

```
